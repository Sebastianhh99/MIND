## Identifique relaciones/patrones encuentra entre los atributos y la posibilidad de tener cáncer, y visceer la posibilidad de no tenerlo.

```python
pd.set_option('display.max_rows', 32)
pd.set_option('display.max_columns', 32)
data.corr()
```

Usando la correlacion de pearson podemos detallar que tenemos una relacion lineal entre los siguientes atributos:
- radius_mean vs perimeter_mean
- radius_mean vs area_mean
- perimeter_mean vs area_mean
- concavity_mean vs concave_points_mean
- radius_se vs perimeter_se
- area_se vs perimeter_se
- radius_worst vs radius_mean
- radius_worst vs perimeter_mean
- radius_worst vs area_mean
Tomando esto en cuenta vemos relacion entre los tamaños medios junto con los tamaños en los que se puede encontrar tamaños malignos
Esto se ve en una tendencia muy creciente entre el tamaño medio y maligno al momento de revisar los datos vemos que si la diferencia entre los tamaños mean y worst es demasiado grande, la tendencia a ser maligno es mayor.

## Realice gráfico de dispersión entre cada par de atributos con la intención de conocer la relación entre ellas y la posibilidad de tener cáncer o no.

```python
for i in range(2,len(data.columns)):
    for j in range(i+1,len(data.columns)):
        data.plot.scatter(x=data.columns[j],y=data.columns[i])
```

En los graficos podemos detallar varias relaciones lineales frente a lo anteriormente dicho sobre los datos acerca de los tamaños y su relacion al momentos de ser benigno o maligno algunos de los datos no tienen ningun tipo de relacion linean visible grafica o numericamente usando la correlacion de pearson.
